```python
import json
coursera_skills_list = None
skills_jobs_map_profiency = None
with open("./coursera/coursera_skills_set.json", "r") as file:
    coursera_skills_list = set(json.load(file))
with open("./coursera/skills_jobs_map_profiency.json", "r") as file:
    skills_jobs_map_profiency = json.load(file)

coursera_skills_list = {skill.lower() for skill in coursera_skills_list}
```

```python
coursera_skills_list
```

```python
missing_job_title_skills = {}
job_title_set = set()
for job_title in skills_jobs_map_profiency:
    job_title_set.add(job_title)
    for skills_profiency in skills_jobs_map_profiency[job_title]:
        if skills_profiency.lower() not in coursera_skills_list:
            if job_title not in missing_job_title_skills:
                missing_job_title_skills[job_title] = []
            missing_job_title_skills[job_title].append(skills_profiency)
```
